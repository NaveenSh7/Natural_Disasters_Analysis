```python
import sqlite3
import pandas as pd
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# --------------------
# 1. Load Data from SQLite
# --------------------
conn = sqlite3.connect("database/disasters.db")
df = pd.read_sql("SELECT * FROM disasters", conn)
conn.close()

# Convert Date column
df["Date"] = pd.to_datetime(df["Date"])

# --------------------
# 2. Create Dashboard Layout
# --------------------
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("Disaster Type Distribution", 
                    "Fatalities by Type",
                    "Economic Loss by Type",
                    "Disaster Timeline"),
    vertical_spacing=0.15, horizontal_spacing=0.1
)

# 2.1 Disaster Type Distribution
type_counts = df["Disaster_Type"].value_counts()
fig.add_trace(
    go.Bar(x=type_counts.index, y=type_counts.values, marker_color="skyblue", name="Count"),
    row=1, col=1
)

# 2.2 Fatalities by Type
fig.add_trace(
    go.Box(x=df["Disaster_Type"], y=df["Fatalities"], boxmean="sd", marker_color="red", name="Fatalities"),
    row=1, col=2
)

# 2.3 Economic Loss by Type
econ_loss = df.groupby("Disaster_Type")["Economic_Loss($)"].sum().sort_values(ascending=False)
fig.add_trace(
    go.Bar(x=econ_loss.index, y=econ_loss.values, marker_color="orange", name="Economic Loss"),
    row=2, col=1
)

# 2.4 Timeline (Fatalities over time by disaster type)
timeline = df.groupby(["Date", "Disaster_Type"])["Fatalities"].sum().reset_index()
for d_type in timeline["Disaster_Type"].unique():
    subset = timeline[timeline["Disaster_Type"] == d_type]
    fig.add_trace(
        go.Scatter(x=subset["Date"], y=subset["Fatalities"], mode="lines", name=d_type),
        row=2, col=2
    )

# --------------------
# 3. Layout Settings
# --------------------
fig.update_layout(
    height=900, width=1200,
    title_text="ðŸŒ Natural Disaster EDA Dashboard (SQLite + Plotly)",
    showlegend=True,
    template="plotly_dark"
)

fig.show()
```

```python
import streamlit as st
import pandas as pd
import plotly.express as px
import sqlite3

# Load data
conn = sqlite3.connect("database/disasters.db")
df = pd.read_sql("SELECT * FROM disasters", conn)
conn.close()
df["Date"] = pd.to_datetime(df["Date"])

# Sidebar filters
st.sidebar.header("Filters")
disaster_types = df["Disaster_Type"].unique()
selected_type = st.sidebar.multiselect("Select Disaster Type", disaster_types, disaster_types)
date_range = st.sidebar.date_input("Select Date Range", [df["Date"].min(), df["Date"].max()])

# Filter data
filtered_df = df[(df["Disaster_Type"].isin(selected_type)) & 
                 (df["Date"] >= pd.to_datetime(date_range[0])) & 
                 (df["Date"] <= pd.to_datetime(date_range[1]))]

# Dashboard layout
st.title("ðŸŒ Natural Disaster Dashboard 2024")
st.subheader("Dataset Preview")
st.dataframe(filtered_df.head())

st.subheader("Disaster Type Distribution")
fig1 = px.histogram(filtered_df, x="Disaster_Type", color="Disaster_Type", template="plotly_dark")
st.plotly_chart(fig1)

st.subheader("Fatalities by Disaster Type")
fig2 = px.box(filtered_df, x="Disaster_Type", y="Fatalities", color="Disaster_Type", template="plotly_dark")
st.plotly_chart(fig2)

st.subheader("Economic Loss by Disaster Type")
fig3 = px.bar(filtered_df.groupby("Disaster_Type")["Economic_Loss($)"].sum().reset_index(),
              x="Disaster_Type", y="Economic_Loss($)", color="Disaster_Type", template="plotly_dark")
st.plotly_chart(fig3)

st.subheader("Fatalities Over Time")
fig4 = px.line(filtered_df, x="Date", y="Fatalities", color="Disaster_Type", template="plotly_dark")
st.plotly_chart(fig4)
```
